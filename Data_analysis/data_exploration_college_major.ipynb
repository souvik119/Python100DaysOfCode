```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
```

```python
df.head()
```

```python
# no of rows and columns in dataframe
df.shape
```

```python
# list the column names
df.columns
```

```python
# remove NaN(Not a Number) values to remove missing/junk data
df.isna()
```

```python
#looks like last row has some NaN values so lets drop that row
# craete a new df with last row removed
clean_df = df.dropna()
```

```python
# access particular column values
clean_df["Starting Median Salary"]
```

```python
#find highest starting salary, chain max method
clean_df["Starting Median Salary"].max()
```

```python
# but which college major earns the max, we need to know the row number or index of above record
clean_df['Starting Median Salary'].idxmax()
```

```python
# so index of row is 43, to see the major that corresponds to that particular row, we can use .loc property
clean_df['Undergraduate Major'].loc[43]
#we can use the .loc property to get entire row if we do not want a specific column value
```

```python
# What college major has the highest mid-career salary? How much do graduates with this major earn?
# clean_df["Mid-Career Median Salary"].max()
print(clean_df['Mid-Career Median Salary'].max())
print(clean_df['Mid-Career Median Salary'].idxmax())
print(clean_df['Undergraduate Major'].loc[8])
```

```python
#Which college major has the lowest starting salary and how much do graduates earn after university?
print(clean_df["Starting Median Salary"].min())
print(clean_df['Undergraduate Major'].loc[clean_df["Starting Median Salary"].idxmin()])
```

```python
# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?
print(clean_df['Mid-Career Median Salary'].min())
print(clean_df['Undergraduate Major'].loc[clean_df["Mid-Career Median Salary"].idxmin()])
```
